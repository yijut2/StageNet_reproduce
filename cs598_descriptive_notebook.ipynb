## Reproducibility summary
This paper introduces StageNet, a novel network model that surpasses existing state-of-the-art models in terms of AUPRC, AUROC, and min(Re, P+) performance on the MIMIC-III dataset. Our experiments primarily focused on identifying optimal parameters to enhance StageNet's performance and analyzing the individual contributions and efficiency of its three-layer architecture.

By replicating the StageNet experiment using identical hyperparameters, we observed a clear superiority of our trained StageNet model compared to both the best weights provided in original paperâ€™s repository and our own trained weights on the MIMIC-III dataset. Furthermore, we conducted a comparative analysis between StageNet and LSTM models to determine if the stage-adaptive convolutional module effectively extracts crucial features from compressed patient information, such as progression patterns. The results revealed that StageNet achieved an impressive AUPRC score of 0.153, surpassing LSTM's score of 0.147.

On the other hand, through multiple rounds of experimentation, we speculate that a dropout rate of 0.3 and a learning rate of 0.001 generally lead to improved performance across the AUPRC, AUROC, and min(Re, P+) metrics.
